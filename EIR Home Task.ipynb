```python
data = pd.read_csv("EIR Home Task - EIR_Home Task.csv")
data = data.drop(columns = ["1"])
data.head()
```

```python
# Convert timestamps to datetime format
data['CREATED_DATE'] = pd.to_datetime(data['CREATED_DATE'])
data['CREATED_DATE minus Hour'] = pd.to_datetime(data['CREATED_DATE minus Hour'])
```

```python
# Calculate average amount per currency
average_per_currency = data.groupby('CURRENCY')['AMOUNT'].mean()

print(average_per_currency)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Resample data by day and calculate total transactions per day
transactions_by_day = data.resample('D', on='CREATED_DATE').size()

# Plot transactions over time (daily)
plt.figure(figsize=(10, 6))
transactions_by_day.plot(kind='line')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.title('Distribution of Transactions Over Time (Daily)')
plt.show()
```

```python
# Resample data by weekly and count transactions
transactions_by_hour = data.resample('W', on='CREATED_DATE').size()

# Plot transactions over time
plt.figure(figsize=(10, 6))
transactions_by_hour.plot(kind='line')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.title('Distribution of Transactions Over Time (Weekly)')
plt.show()
```

```python
# Resample data by day and calculate total transactions per day
transactions_by_day = data.resample('Q', on='CREATED_DATE').size()

# Plot transactions over time (daily)
plt.figure(figsize=(10, 6))
transactions_by_day.plot(kind='line')
plt.xlabel('Quarter')
plt.ylabel('Number of Transactions')
plt.title('Distribution of Quarterly Transactions Over Time')
plt.show()
```

```python
# Trends or patterns in transaction activity for specific currencies

# Group by Date and CURRENCY, and calculate TransactionCount
currency_trends = data.groupby([data['CREATED_DATE'].dt.date, 'CURRENCY'])['TRANSACTION_ID'].count().reset_index()

# Rename columns for clarity
currency_trends.rename(columns={'CREATED_DATE': 'Date', 'CURRENCY': 'Currency', 'TRANSACTION_ID': 'TransactionCount'}, 
                       inplace=True)

# Sort by Date
currency_trends.sort_values(by='Date', inplace=True)
```

```python
currency_trends
```

```python
# Find the row with the maximum TransactionCount
max_transaction_day_row = currency_trends.loc[currency_trends['TransactionCount'].idxmax()]

# Extract the Date and TransactionCount values
max_transaction_day = max_transaction_day_row['Date']
max_transaction_count = max_transaction_day_row['TransactionCount']

print(f"Maximum Transaction Day: {max_transaction_day}")
print(f"Maximum Transaction Count: {max_transaction_count}")
```

```python
# Find the row with the minimum TransactionCount
min_transaction_day_row = currency_trends.loc[currency_trends['TransactionCount'].idxmin()]

# Extract the Date and TransactionCount values
min_transaction_day = min_transaction_day_row['Date']
min_transaction_count = min_transaction_day_row['TransactionCount']

print(f"Minimum Transaction Day: {min_transaction_day}")
print(f"Minimum Transaction Count: {min_transaction_count}")
```
